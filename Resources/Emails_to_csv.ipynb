```python
def email_analyse(inputfile):
    with open(inputfile,"r") as f:
        data=f.read()
    global df
    email = Parser().parsestr(data)
    email_to = list()
    to_email_list = list()
    email_from = ""
    email_body = ""
    email_subject = ""
    if email["to"]:
        

        to_email_list = email['to']
        to_email_list = to_email_list.replace("\n","")
        to_email_list = to_email_list.replace("\t","")
        to_email_list = to_email_list.replace(" ","")

        to_email_list = to_email_list.split(",")

        for i in to_email_list:
            email_to.append(i)

        email_from = email["from"]
        email_body = email.get_payload()
        email_body = email_body.replace("\n","")
        email_body = email_body.replace("\t","")

        email['subject']
        email_subject = email_subject.replace("\n","")
        email_subject = email_subject.replace("\t","")

        return {"To":email_to,"From":email_from,"Subject":email_subject,"Body":email_body}
```

```python
temp = dict()
for directory, subdirectory,filenames in os.walk(rootdir):
    for filename in  filenames:
        try:
            temp = email_analyse(os.path.join(directory,filename))
            temp["account_name"] = directory
            # print(temp)
            df_list.append(temp)
        except:
            print(os.path.join(directory,filename))
```
